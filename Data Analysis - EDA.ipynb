# Find popular genres: 
You can determine which genres are popular by looking at columns like Movies Released or Tickets Sold. Analyzing these numbers will give you insights into what types of movies attract more audiences.

The most popular genre by year in terms of tickets sold.

```python
most_sold_by_year = filmGenre.groupby("Year").apply(lambda df: df.loc[df["Tickets Sold"].idxmax()])
most_sold_by_year = most_sold_by_year.set_index("Year")
most_sold_by_year
```

```python
pop_genres = filmGenre.set_index("Year").groupby("Genre").apply(lambda df: df[["Movies Released", "Tickets Sold", "Inflation-Adjusted Gross"]].sort_values(by="Year"))
pop_genres
```

```python
pivoted_df = pop_genres.reset_index()
pivoted_df = pivoted_df.pivot_table(values=["Movies Released", "Tickets Sold", "Inflation-Adjusted Gross"], index="Genre", aggfunc="sum")
pivoted_df
```

```python
plt.figure(figsize=(16, 8))
plt.title("Movies Released by Genre from 1995 to 2018")
sns.barplot(data=pivoted_df, x=pivoted_df.index, y=pivoted_df["Movies Released"])
plt.xticks(rotation=45)
```

```python
filtered_df = filmGenre[["Genre", "Year", "Movies Released"]]
filtered_df
```

```python
plt.figure(figsize=(16, 8))
sns.lineplot(data=filtered_df, x=filtered_df['Year'], y=filtered_df['Movies Released'], hue=filtered_df["Genre"])
```
